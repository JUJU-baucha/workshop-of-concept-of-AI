# **3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:**
Based on the dataset Answer the following question:

**What percent of the total fare revenue came from each passenger class?**


To answer the question perform following operation:

1. Calculate the total Fare paid across all passengers in the Titanic dataset.

```python
total_fare = df['Fare'].sum()
total_fare
```

2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.

```python
first_class = df[df['Pclass']==1]
first_class_fare = first_class['Fare'].sum()
first_class_fare
```

3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).

```python
second_class = df[df['Pclass']==2]
second_class_fare = second_class['Fare'].sum()
second_class_fare
```

```python
third_class = df[df['Pclass']==3]
third_class_fare = third_class['Fare'].sum()
third_class_fare
```

4. Combine the fare totals from first, second, and third classes into a list.

```python
fare_totals = ['first_class_fare', 'second_class_fare', 'third_class_fare']
fare_totals
```

5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.

```python
total_fare = df['Fare'].sum()
first_class_proportion = first_class_fare / total_fare
second_class_proportion = second_class_fare / total_fare
third_class_proportion = third_class_fare / total_fare
print("Proportion of fare revenue by class:")
print("First Class:", first_class_proportion)
print("Second Class:", second_class_proportion)
print("Third Class:", third_class_proportion)
```

1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).


```python
new_column = []
for age in df['Age']:
  if age<18:
    new_column.append('child')
  elif age>= 18 and age==64:
    new_column.append('adult')
  else:
    new_column.append('senior')
df['age_group'] = new_column
df.head()
```

```python
total_number = df['PassengerId'].count()
total_number
```

```python
total_passengers_age= df['age_group'].value_counts()
total_passengers_age
```

```python
proportion = df['age_group'].value_counts() / total_number
proportion
```

```python
percentage = proportion *100
percentage
```
